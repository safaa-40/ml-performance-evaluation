```python
import numpy as np

def calculate_preference_selection_index(models_data, weights):
    # Extract criteria (accuracy, precision, training time, execution time)
    criteria = models_data[:, 1:].astype(float)

    # Normalize data
    normalized_criteria = (criteria - np.min(criteria, axis=0)) / (np.max(criteria, axis=0) - np.min(criteria, axis=0))

    # Multiply each criterion by its weight
    weighted_criteria = normalized_criteria * weights

    # Calculate Preference Selection Index for each model
    preference_selection_index = np.sum(weighted_criteria, axis=1)

    return preference_selection_index

# Define weights for each criterion (accuracy, precision, training time, execution time)
weights = np.array([1, 1, 1, 1])

# Calculate Preference Selection Index for each model
preference_selection_index = calculate_preference_selection_index(config_matrix1, weights)

samples = [(1000, 26), (1000, 20), (1000, 10), (2500, 26), (2500, 20), (2500, 10), (5000, 26), (5000, 20), (5000, 10)]

# Rank models based on Preference Selection Index
ranked_models_indices = np.argsort(preference_selection_index)[::-1]

# Print ranked models
for i, idx in enumerate(ranked_models_indices):
    print(f"Rank {i+1}: {config_matrix1[idx][0]} (Preference Selection Index: {preference_selection_index[idx]})")
```
